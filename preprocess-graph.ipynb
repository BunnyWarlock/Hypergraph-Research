```python
import networkx as nx
```

```python
G = nx.read_edgelist('bio-diseasome.edges', edgetype=float)
```

```python
print('Edges after removing multi-edges:',len(G.edges()))
```

```python
i=0
for u, v in G.edges():
    if u == v:
        G.remove_edge(u,v)
        i = i+1
print('Total no. of self loops:',i)
print('Edges after removing self loops:', len(G.edges()))
```

```python
nx.write_edgelist(G, path='bio-diseasome-processed.edges', data=False)
```
